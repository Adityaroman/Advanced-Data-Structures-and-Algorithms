```python
Here's the explanation in a single table format for both **Min-Priority Queue** and **Max-Priority Queue**, including how tie-breaking works:

| **Task** | **Priority** | **Min-Priority Queue (Low to High)** | **Max-Priority Queue (High to Low)** |
|----------|--------------|--------------------------------------|--------------------------------------|
| A        | 5            | 5th (last)                           | 1st (first)                          |
| B        | 3            | 3rd                                  | 3rd                                  |
| C        | 1            | 1st (first)                          | 5th (last)                           |
| D        | 2            | 2nd                                  | 4th                                  |
| E        | 4            | 4th                                  | 2nd                                  |
|          |              | **Order: C D B E A**                 | **Order: A E B D C**                 |

### Note:
- If two tasks have the **same priority**, the one that was **inserted first** will come first in the queue.
```

```python
#max_Priority implementation (Exam)
class Node:
    def __init__(self, NewData, NewPriority):  # Corrected __init__
        self.Data = NewData
        self.Priority = NewPriority

class PriorityQueue:
    def __init__(self):  # Corrected __init__
        self.items = []

    def isempty(self):
        return self.items == []

    def enqueue(self, item):
        for i in range(len(self.items)):
            if item.Priority >= self.items[i].Priority:
                self.items.insert(i, item)
                break
        else:
            self.items.append(item)

    def dequeue(self):
        if not self.isempty():
            return self.items.pop(0)

    def peek(self):
        if not self.isempty():
            return self.items[0]

    def size(self):
        return len(self.items)

    def display(self):
        for item in self.items:
            print('|', format(str(item.Data) + "(" + str(item.Priority) + ")", '>6'), end=" ")
            print("|")

q = PriorityQueue()

while True:
    print("\nProgram to implement Max-Priority Queue")
    print("1. Enqueue")
    print("2. Dequeue")
    print("3. Peek")
    print("4. Display")
    print("5. Exit")

    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print("Invalid input. Please enter a number.")
        continue

    if choice == 1:
        num = input("Enter the task name or number: ")
        prio = int(input("Enter the priority value: "))
        newnode = Node(num, prio)
        q.enqueue(newnode)

    elif choice == 2:
        if q.isempty():
            print("Queue is empty")
        else:
            popnode = q.dequeue()
            print("Item dequeued:", popnode.Data, "with priority", popnode.Priority)

    elif choice == 3:
        if q.isempty():
            print("Queue Underflow!")
        else:
            topnode = q.peek()
            print("Item at front:", topnode.Data, "with priority", topnode.Priority)

    elif choice == 4:
        if q.isempty():
            print("Queue is empty")
        else:
            q.display()

    elif choice == 5:
        print("Quitting...")
        break

    else:
        print("Invalid choice! Please enter the correct choice.")
```
