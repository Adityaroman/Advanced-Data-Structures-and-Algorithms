### **Merge Sort Algorithm**
1. **Divide:**  
   - If the list has more than one element:  
     - Find the middle of the list.  
     - Divide the list into two halves.  

2. **Conquer:**  
   - Recursively apply merge sort on both halves.  

3. **Combine:**  
   - Merge the two sorted halves using the `merge()` function.  

---

### **Merge Function Algorithm**
**merge(list, lstart, lend, rstart, rend):**  
1. Create an empty list named `temp`.  
2. Set `i = lstart`.  
3. Set `j = rstart`.  
4. While `i <= lend` and `j <= rend`, do:  
   - If `list[i] <= list[j]`:  
     - Append `list[i]` into `temp`.  
     - Set `i = i + 1`.  
   - Else:  
     - Append `list[j]` into `temp`.  
     - Set `j = j + 1`.  
5. While `i <= lend`, do:  
   - Append `list[i]` into `temp`.  
   - Set `i = i + 1`.  
6. While `j <= rend`, do:  
   - Append `list[j]` into `temp`.  
   - Set `j = j + 1`.  
7. Copy elements from `temp` back to the original list from `lstart` to `rend`.  


```python
# Function to merge two sorted subarrays
def merge(arr, lstart, lend, rstart, rend):
    temp = []  # Create an empty list
    i, j = lstart, rstart
    
    # Merge elements from both subarrays
    while i <= lend and j <= rend:
        if arr[i] <= arr[j]:
            temp.append(arr[i])
            i += 1
        else:
            temp.append(arr[j])
            j += 1
    
    # Add remaining elements from left subarray
    while i <= lend:
        temp.append(arr[i])
        i += 1
    
    # Add remaining elements from right subarray
    while j <= rend:
        temp.append(arr[j])
        j += 1
    
    # Copy merged elements back to the original array
    for k in range(len(temp)):
        arr[lstart + k] = temp[k]

# Function to perform merge sort
def merge_sort(arr, left, right):
    if left < right:
        mid = (left + right) // 2
        
        # Recursively divide and sort
        merge_sort(arr, left, mid)
        merge_sort(arr, mid + 1, right)
        
        # Merge sorted subarrays
        merge(arr, left, mid, mid + 1, right)

# Example usage
arr = [38, 27, 43, 3, 9, 82, 10]
merge_sort(arr, 0, len(arr) - 1)
print(arr)  # Output: [3, 9, 10, 27, 38, 43, 82]
```
